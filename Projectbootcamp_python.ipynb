```python
import pandas as pd
```

```python
df1 = pd.read_csv('/content/Finance_1.csv')
```

```python
df2 = pd.read_csv('/content/Finance_2.csv')
```

```python
df1['issue_d_parsed'] = pd.to_datetime(df1['issue_d'], format='%b-%y', errors='coerce')
```

```python
df1['loan_year'] = df1['issue_d_parsed'].dt.year
```

```python
yearly_loan_stats = df1.groupby('loan_year') [['loan_amnt']].sum().reset_index()
```

[] # YEAR WISE LOAN AMONT STATS

```python
print("year-wise total loan amount:")
print(yearly_loan_stats)
```

[ ] # GRADE AND SUB-GRADE WISE REVOL_BAL

```python


df1.columns = df1.columns.str.strip()
df2.columns = df2.columns.str.strip()

merged_df = pd.merge(df1, df2, on='id', how='inner')
grade_revol_stats = merged_df.groupby(['grade', 'sub_grade'])['revol_bal'].sum().reset_index()

print(grade_revol_stats)
```

[ ] # TOTAL PAYMENT FOR VERIFIED STATUS VS TOTAL PAYMENT FOR NON-VERIFIED STATUS


```python

merged_df = pd.merge(df1, df2, on='id', how='inner')

verification_payment = merged_df.groupby('verification_status')['total_pymnt'].sum().reset_index()
print(verification_payment)
```

[ ] # STATE WISE AND MONTH WISE LOAN STATUS

```python

df1['issue_month'] = pd.to_datetime(df1['issue_d'], format='%b-%y', errors='coerce')

df1['loan_month'] = df1['issue_month'].dt.strftime('%b-%Y')

loan_stats = df1.groupby(['addr_state', 'loan_month', 'loan_status']).size().reset_index(name='total_loans')

print(loan_stats)
```
